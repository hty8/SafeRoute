```python
import pandas as pd
```

```python
df = pd.read_csv("crime_data.csv")
```

```python
df = df[['Latitude','Longitude']]
```

```python
df.head()
```

```python
df = df.dropna()
```

```python
df.dtypes
```

```python
df['Latitude'] = df['Latitude'].astype(str)
df['Longitude'] = df['Longitude'].astype(str)
df
```

```python
f = lambda x: x["Latitude"][:7]
g = lambda x: x["Longitude"][:8]
df["Latitude"] = df.apply(f, axis=1)
df["Longitude"] = df.apply(g, axis=1)
df
```

```python
df['Location'] = df['Latitude'].map(str) + ', ' + df['Longitude']
df
```

```python
df.to_csv('crimes_simplified_location.csv')
```

```python
df['Location'].value_counts()
```

```python
df_counts = df.Location.value_counts().reset_index().rename(columns={'index': 'Location', 'Location': 'count'})
df_counts
```

```python
df_counts.head()
```

```python
df_counts
```

```python
df_counts[['lat','lon']] = df_counts.Location.apply(lambda x: pd.Series(str(x).split(", ")))
df_counts
```

```python
df_counts.to_csv('crimes_simplified_location_count_w_lat_lon.csv')
```
